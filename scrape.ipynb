I used JS within the browser to get all the links which I have saved in `./links.txt`.
An issue with the library website is that is dynamically loads content as you scroll so this was the easiest way to bring it into python without emulating a whole browser.

```
function links(){
  let l = Array
            .from( document.querySelectorAll('a') )
            .filter( i => i.href.includes('objects/PC020402') )
            .map( i => i.href );
  return '[' + links.join(',') + ']';
}
```

````python
links = open('./links.txt').read()
links = links.split(',')

print("There are {} links".format(len(links)))
````

Download links for each thesis document are only available on their "detail view" webpage. This next block collects all of those links and creates a list of tuples.

````python
# COLLECT DOWNLOAD LINKS
import requests
from bs4 import BeautifulSoup

def get_download_link( url ):
    res = requests.get(url);
    soup = BeautifulSoup(res.text, 'html.parser')
    link = soup.find('a', href=True, attrs={'class': 'dlButton'})
    # format download link
    dl_link = "https://digitalarchives.library.newschool.edu" + link.get('href')
    return dl_link
# ---------------------------
download_links = []

print("Collecting {} links".format(len(links)))

for link in links:
    item = (link[80:], get_download_link(link))
    download_links.append(item)
````

````python
from IPython.core.display import HTML

print("{} download links available".format(len(download_links)))

print("\nFirst 40 links belong to:")

s = """<ul>"""
for d in download_links[0:40]:
    id, url = d
    s += """<li><a href="{}">{}</a>""".format(url, id)
s += """</ul>"""

display(HTML(s))
````

````python
# SAVE PDF FILES
import requests
import os.path

def save_pdf( link, file_path ):
    pdf_file = requests.get(link)
    with open(file_path,'wb') as output:
      output.write(pdf_file.content)
      print("Saved: {}".format(file_path))
        
for link in download_links:
    # extract details
    filename, url = link
    # create filepath
    filepath = "./PDFs/{}.pdf".format(filename)
    
    # if file exists, skip
    # otherwise, write file
    if os.path.isfile(filepath):
        print("{} already exists".format(filepath))
    else:
        save_pdf(url,filepath)
````
